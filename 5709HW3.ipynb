```python
import matplotlib.pyplot as plt
import numpy as np
import sympy
# 这两行代码使得 pyplot 画出的图形中可以显示中文
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
# 生成数据
r = np.arange(1, 10, 1)
b = np.log(0.95)/np.log(1-(0.8)**r)
r1 = np.arange(1, 10, 1)
b1 = np.log(0.95)/np.log(1-(0.3)**r1)

dist = (b-b1)**2
i = np.argmin(dist)
print(r[i], b[i])
print(r1[i], b1[i])
# print("s=0.8")
# print(1-(1-0.8**4)**5)
# print(1-(1-0.8**4)**6)
# print(1-(1-0.8**3)**5)
# print(1-(1-0.8**3)**6)
# print("s=0.3")
# print(1-(1-0.3**4)**5)
# print(1-(1-0.3**4)**6)
# print(1-(1-0.3**3)**5)
# print(1-(1-0.3**3)**6)
# r = 4 b =5
# r = 3 b =5
# r = 4 b =6
```

```python

#[s1,s2]=sympy.solve('1-(1-0.8**x)**y=0.95','1-(1-0.3**x)**y=0.0.05');#%求方程组的解

plt.plot(r, b, 'go:', label='', linewidth=2,color='r') # 颜色绿色，点形圆形，线性虚线，设置图例显示内容，线条宽度为2
plt.plot(r1, b1, 'go:', label='', linewidth=2,color='b') # 颜色绿色，点形圆形，线性虚线，设置图例显示内容，线条宽度为2
plt.ylabel('b') # 横坐标轴的标题
plt.xlabel('r') # 纵坐标轴的标题
# plt.xticks(np.arange(0, 11, 1)) # 设置横坐标轴的刻度为 0 到 10 的数组
# plt.ylim([-2, 2]) # 设置纵坐标轴范围为 -2 到 2
#plt.legend() # 显示图例, 图例中内容由 label 定义
# plt.grid() # 显示网格
# plt.title('我的第一个 Python 图形') # 图形的标题

# 显示图形
plt.show()
```

```python
sim = np.arange(0, 1.1, 0.1)

for i in range(1,10):
    for j in range(1,10):
        pro = 1 - (1-sim**i)**j
        plt.plot(sim, pro, label='', color='r') # 颜色绿色，点形圆形，线性虚线，设置图例显示内容，线条宽度为2

plt.ylabel('pro') # 横坐标轴的标题
plt.xlabel('sim') # 纵坐标轴的标题
plt.axhline(y=0.95,ls=":",c="blue")#添加水平直线
plt.axvline(x=0.8,ls=":",c="blue")#添加垂直直线
plt.axhline(y=0.05,ls=":",c="blue")#添加水平直线
plt.axvline(x=0.3,ls=":",c="blue")#添加垂直直线
plt.plot(0.8,0.95,color='black',marker="*")  #markevery是要出现标志的位置的点，一般会是一个list
plt.plot(0.3,0.05,color='black',marker="*")  #markevery是要出现标志的位置的点，一般会是一个list
plt.show()
```
