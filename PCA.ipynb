```python
from numpy import *
import matplotlib.pyplot as plt
X = mat([[7,4,3,4],[4,1,8,3],[6,3,5,2],[8,3,2,10],[4,5,0,9],[1,3,2,5],[6,6,3,2],[8,3,3,6]])
#将矩阵中心化，即减去均值：
X_new = X - mean(X, axis=0)
```

```python
X_new
```

```python
C = dot(X_new.T, X_new) / (X_new.shape[0] - 1)
```

```python
eig_vals, eig_vecs = linalg.eig(C)
```

```python
eig_vecs.shape
```

```python
X_new.shape
```

```python
eig_vecs
```

```python
eig_vals
```

```python
e2k = eig_vecs[:,[0,2]]
X_pca = dot(X_new, e2k)
X_pca
```

```python
X_pca.shape
```

```python
PCA_array = X_pca.A
```

```python
plt.figure()
for i in PCA_array:
    plt.scatter(i[0],i[1],alpha=.7)
plt.legend()
plt.title('Numerical Example of PCA')
plt.show()
```
